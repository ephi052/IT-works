```python
import pandas as pd
df = pd.read_csv('/content/formated_list.csv')
df = df.fillna('None')
df
```

```python
"""
This program will create a script for adding accses to a list of accounts.

Parameters
----------
accunt_from_list : list
    A list of accounts to give accses to.
accunt_to_list : list
    A list of accounts to give accses from.
sellected_accsess_type : str
    The type of accses to give.
add_hebrew_line : bool
    If True, the script will add a line for the hebrew calendar.
"""
def check_data():
  return len(df.iloc[0]) % 2 == 0  

class get_list_of_lists():
  """
  This function takes a dataframe with a list of lists in each row and returns two lists of lists.
  The first list of lists contains the first half of the lists in the dataframe.
  The second list of lists contains the second half of the lists in the dataframe.
  """
  def __init__(self) -> None:
    number_of_lists_in_list_of_lists = len(df.iloc[0])/2
    self.from_list_of_lists = []
    self.to_list_of_lists = []
    for list_in_list_of_lists in range(int(number_of_lists_in_list_of_lists)):
      column = df["give_" + str(list_in_list_of_lists+1)].tolist()
      clean_column = [i for i in column if i != 'None']
      self.from_list_of_lists.append(clean_column)
      column = df["access_to_" + str(list_in_list_of_lists+1)].tolist()
      clean_column = [i for i in column if i != 'None']
      self.to_list_of_lists.append(clean_column)
  
  def from_list_of_lists(self):
    return self.from_list_of_lists
  
  def to_list_of_lists(self):
    return self.to_list_of_lists

def check_lists(from_list_of_lists,to_list_of_lists):
  """
  Checks if the two lists of lists have the same length.

  Parameters
  ----------
  from_list_of_lists : list of lists
      The list of lists to be checked.
  to_list_of_lists : list of lists
      The list of lists to be checked.

  Returns
  -------
  bool
      True if the two lists of lists have the same length, False otherwise.
  """
  return len(from_list_of_lists) == len(to_list_of_lists)

def sellect_permission_type():
  """
  This function is used to select the type of access rights for the user.
  
  Returns
  -------
  sellected_accsess_type : str
      The type of access rights for the user.
  """
  accsess_lang_types = ('Reviewer','Editor')
  select = int (input('\n Sellct Type of Access Rights \n Viewer = 0, Editor = 1 : \n >'))
  sellected_accsess_type = accsess_lang_types [select]
  return sellected_accsess_type

def add_hebrew():
  """
  This function asks the user if he wants to generate a Hebrew calendar.

  Returns
  -------
  True if the user wants to generate a Hebrew calendar, False otherwise.
  """
  add_hebrew_line = input('\n do you whant to genorate for hebrow caleners also? \n y (yes) : \n >')
  return add_hebrew_line == 'y'

def make_scripts(from_list_of_lists,to_list_of_lists,sellected_accsess_type,add_hebrew_line):
  """
  This function makes a script for each list in the list of lists.

  Parameters
  ----------
  from_list_of_lists : list of lists
      A list of lists of strings. Each string is a file name.
  to_list_of_lists : list of lists
      A list of lists of strings. Each string is a file name.
  sellected_accsess_type : string
      A string that is the accsess type.
  add_hebrew_line : boolean
      A boolean that is True if the script should add a hebrew line.

  Returns
  -------
  None
  """
  for i in range(len(from_list_of_lists)):
    list_from = from_list_of_lists[i]
    list_to = to_list_of_lists[i]    
    make_script(list_from,list_to,sellected_accsess_type,add_hebrew_line)

def make_script(accunt_from_list,accunt_to_list,sellected_accsess_type,add_hebrew_line):
  """
  This function will create a script for adding accses to a list of accounts.

  Parameters
  ----------
  accunt_from_list : list
      A list of accounts to give accses to.
  accunt_to_list : list
      A list of accounts to give accses from.
  sellected_accsess_type : str
      The type of accses to give.
  add_hebrew_line : bool
      If True, the script will add a line for the hebrew calendar.

  Returns
  -------
  None
  """  
  STATIC_STR_1 = "-MailboxFolderPermission"
  STATIC_STR_2 = ' -identity “'
  STATIC_STR_3 = " -user “"
  STATIC_STR_4 = "” -AccessRights "
  OPERATION_TYPE = "Add"
  SELLECTED_LANG_TYPEE = ':\Calendar”'
  SELLECTED_LANG_TYPEH = ':\לוח שנה”'  
  print(accunt_to_list,' --> will get accses to --> ',accunt_from_list)  
  with open("scripts.txt","a",encoding="utf-8") as file:
    for i in range (len(accunt_from_list)):
      account_to_delegate_from = accunt_from_list[i]
      for r in range (len(accunt_to_list)):
        account_to_delegate_to = accunt_to_list[r]
        script_line = [OPERATION_TYPE.title()+ STATIC_STR_1 + STATIC_STR_2 +
                account_to_delegate_from + SELLECTED_LANG_TYPEE + STATIC_STR_3 +
                account_to_delegate_to + STATIC_STR_4 + sellected_accsess_type]
        script_line = str(script_line).join(map(str,script_line))
        print(script_line)
        file.write(script_line)
        file.write('\n')
        if add_hebrew_line:
          script_line_hebrew = [OPERATION_TYPE.title()+ STATIC_STR_1 + STATIC_STR_2 +
                  account_to_delegate_from + SELLECTED_LANG_TYPEH + STATIC_STR_3 +
                  account_to_delegate_to + STATIC_STR_4 + sellected_accsess_type]
          script_line_hebrew = str(script_line_hebrew).join(map(str,script_line_hebrew))
          print(script_line_hebrew)        
          file.write(script_line_hebrew)
          file.write('\n')
```

```python
if __name__ == '__main__':
  if check_data():
    if check_lists(get_list_of_lists().to_list_of_lists,get_list_of_lists().from_list_of_lists):
      make_scripts(get_list_of_lists().to_list_of_lists,get_list_of_lists().from_list_of_lists,sellect_permission_type(),add_hebrew())
    else:
      print('problem with list of lists')
      print(get_list_of_lists().to_list_of_lists,get_list_of_lists().from_list_of_lists)
  else:
    print('Data format is not correct \n the correct format is')
    print(pd.DataFrame(data={'give_1': ['user', 'user'], 'access_to_1': [ 'user', 'user'],'give_2': ['user', 'user'], 'access_to_2': [ 'user', 'user']}))
```
